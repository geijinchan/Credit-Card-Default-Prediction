## 3.2 Probability of defaulting payment next month

```python
pay_count = df1['default.payment.next.month'].value_counts(normalize=True)*100
pay_count.plot.bar(figsize=(6,6))
plt.xticks(fontsize=12, rotation=0)
plt.yticks(fontsize=12)
plt.title("Probability Of Defaulting Payment Next Month", fontsize=15)
for x,y in zip([0,1],pay_count):
    plt.text(x,y,y,fontsize=12)
plt.show()
```

## 3.3 Age catagory of client

```python
df1['AGE'].count()
bins = [20,30,40,50,60,70,80]
names = ['21-30','31-40','41-50','51-60','61-70','71-80']
df1['Age_Bins'] = pd.cut(x=df1.AGE,bins=bins,labels=names,right=True)

age_0 = (df1.Age_Bins[df1['default.payment.next.month']==1].value_counts())
age_1 = (df1.Age_Bins[df1['default.payment.next.month']==0].value_counts())

plt.subplots(figsize=(8,5))

plt.bar(age_0.index, age_0.values,label='0')
plt.bar(age_1.index, age_1.values,label='1')
for x,y in zip(names,age_1):
    plt.text(x,y,y,fontsize=12)
for x,y in zip(names,age_0):
    plt.text(x,y,y,fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.title("Number of clients in each age group", fontsize=15)
plt.legend(loc='upper right', fontsize=15)
plt.show()
```

Age group of 21-30 have maximum clients

# 4. Model Building

## 4.1 Splitting data

```python
df1_X = df1.drop(['default.payment.next.month','Age_Bins'], axis=1)
df1_Y = df1['default.payment.next.month']

X_train, X_test, y_train, y_test = train_test_split(df1_X, df1_Y, test_size=0.2, random_state=10)
```

## 4.2 Model training

```python
Model_LR = LogisticRegression()
Model_LR.fit(X_train, y_train)

y_pred = Model_LR.predict(X_test)

print(classification_report(y_pred, y_test))
print(confusion_matrix(y_pred, y_test))
print('\nAccuracy Score for Logistic Regression: ', accuracy_score(y_pred,y_test))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Create a RandomForestClassifier model
rfm = RandomForestClassifier(
    n_estimators=20,
    oob_score=True,
    n_jobs=1,
    random_state=42,
    max_features=None,
    min_samples_leaf=10
)

# Train the model
rfm.fit(X_train, y_train)

# Make predictions
rfm_y_pred = rfm.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(rfm_y_pred, y_test)
print("Accuracy Score: ", accuracy)
```

## 4.3 Model Evaluation

- Accuracy Score for Logistic Regression:  0.7801666666666667
- Accuracy Score for Random Forest:  0.8256666666666667

## 4.4 Choosing model

Sine Random Forest is giving more accuracy so model will be used is Random Forest

## 4.5 Hypertuning 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Create a RandomForestClassifier model
rfm = RandomForestClassifier(random_state=42)

# Define the hyperparameter grid
param_grid = {
    'n_estimators': [10, 20, 30],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['auto', 'sqrt']
}

# Perform Grid Search to find the best hyperparameters
grid_search = GridSearchCV(rfm, param_grid, cv=5)
grid_search.fit(X_train, y_train)

# Get the best model and its hyperparameters
best_model = grid_search.best_estimator_
best_params = grid_search.best_params_

# Make predictions using the best model
rfm_y_pred = best_model.predict(X_test)

# Calculate and print the accuracy score
accuracy = accuracy_score(rfm_y_pred, y_test)
print("Accuracy Score: ", accuracy)
print("Best Hyperparameters: ", best_params)
```

## 4.6 Saving model

```python
with open('best_model.pkl', 'wb') as f:
    pickle.dump(best_model, f)
```
